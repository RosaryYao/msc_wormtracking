## FGBG model

```python
import matplotlib.pyplot as plt
```

```python
m1_train_accuracy = [0.9358, 0.9719, 0.9797]
m2_train_accuracy = [0.9521, 0.9742, 0.9801]
```

```python
m1_val_accuracy = [0.5238, 0.9661, 0.9921]
m2_val_accuracy = [0.3529, 0.9714, 0.9954]
```

```python
m1_train_loss = [0.1975, 0.1218, 0.1098]
m2_train_loss = [0.1768, 0.1224, 0.1038]
```

```python
m1_val_loss = [0.6789, 0.271, 0.2539]
m2_val_loss = [1.5041, 0.2386, 0.2888]
```

```python
total_train_loss = [0.3925, 0.2625, 0.2319]
total_val_loss = [2.2013, 0.5279, 0.561]
```

```python
import matplotlib.font_manager as font_manager
from matplotlib.pyplot import figure

font = font_manager.FontProperties(family='serif',
                                   style='normal', size=11)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

plt.plot([1,2,3], m1_train_accuracy, color='tab:blue', label="Model 1 training accuracy")
plt.plot([1,2,3], m2_train_accuracy, color='tab:orange', label="Model 2 training accuracy")
plt.plot([1,2,3], m1_val_accuracy, color='tab:green', label="Model 1 validation accuracy")
plt.plot([1,2,3], m2_val_accuracy, color='tab:red', label="Model 2 validation accuracy")
plt.xlim(0.8, 3.2)
plt.ylim(0.3,1.03)

csfont = {'fontname':'serif'}
hfont = {'fontname':'serif'}

plt.title('Accuracy of FGBF Model 1 and 2',**csfont)
plt.xlabel('Epoch', **hfont, fontsize=12)
plt.ylabel('Accuracy', **hfont, fontsize=12)
plt.legend(prop=font)
plt.xticks([1,2,3], **hfont)
plt.legend(prop=font, bbox_to_anchor=(1.04,1), borderaxespad=0)

#ax.set_aspect(1.0/ax.get_data_ratio(), adjustable='box')
plt.savefig('pics/results/fgbg_accuracy.png')

figure(figsize=(8, 6), dpi=80)
plt.show()
```

```python
import matplotlib.font_manager as font_manager
from matplotlib.pyplot import figure

font = font_manager.FontProperties(family='serif',
                                   style='normal', size=11)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

plt.plot([1,2,3], m1_train_loss, color='tab:blue', label="Model 1 training loss")
plt.plot([1,2,3], m2_train_loss, color='tab:orange', label="Model 2 training loss")
plt.plot([1,2,3], m1_val_loss, color='tab:green', label="Model 1 validation loss")
plt.plot([1,2,3], m2_val_loss, color='tab:red', label="Model 2 validation loss")
plt.plot([1,2,3], total_train_loss, color='tab:purple', label="Training total loss")
plt.plot([1,2,3], total_val_loss, color='tab:cyan', label="Validation  total loss")
plt.xlim(0.8, 3.2)
plt.ylim(0,2.5)

csfont = {'fontname':'serif'}
hfont = {'fontname':'serif'}

plt.title('Loss of FGBF Model 1 and 2',**csfont)
plt.xlabel('Epoch', **hfont, fontsize=12)
plt.ylabel('Loss', **hfont, fontsize=12)
plt.legend(prop=font)
plt.xticks([1,2,3], **hfont)
plt.legend(prop=font, bbox_to_anchor=(1.04,1), borderaxespad=0)

#ax.set_aspect(1.0/ax.get_data_ratio(), adjustable='box')
plt.savefig('pics/results/fgbg_loss.png')

figure(figsize=(8, 6), dpi=80)
plt.show()
```
